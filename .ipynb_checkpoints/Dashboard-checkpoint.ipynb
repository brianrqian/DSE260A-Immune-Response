```python
import dash
from dash import html, dcc
import plotly.express as px
import dash_table
import pandas as pd
import dash_bootstrap_components as dbc
import plotly.graph_objects as go
```

```python
titer_df = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/prediction_data/2022BD_plasma_ab_titer.tsv", sep='\t')
cell_freq = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/prediction_data/2022BD_pbmc_cell_frequency.tsv", sep='\t')
gene = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/prediction_data/2022BD_pbmc_gene_expression.tsv", sep='\t')
```

```python
df_2020_cell_freq = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2020LD_pbmc_cell_frequency.tsv", sep='\t')
df_2021_cell_freq = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2021LD_pbmc_cell_frequency.tsv", sep='\t')
df_2020_gene = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2020LD_pbmc_gene_expression.tsv", sep='\t')
df_2021_gene = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2021LD_pbmc_gene_expression.tsv", sep='\t')
df_2020_titer = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2020LD_plasma_ab_titer.tsv", sep='\t')
df_2021_titer = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2021LD_plasma_ab_titer.tsv", sep='\t')
```

```python
df_2020_specimen = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2020LD_specimen.tsv", sep='\t')
df_2020_subject = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2020LD_subject.tsv", sep='\t')
df_2021_specimen = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2021LD_specimen.tsv", sep='\t')
df_2021_subject = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/training_data/2021LD_subject.tsv", sep='\t')
```

```python
titer_df = pd.concat([df_2020_titer, df_2021_titer])
cell_freq = pd.concat([df_2020_cell_freq, df_2021_cell_freq])
gene = pd.concat([df_2020_gene, df_2021_gene])
```

```python
#titer_df = pd.read_csv('titer.csv')

#df_2022_specimen = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/prediction_data/2022BD_specimen.tsv", sep='\t')
#df_2022_subject = pd.read_csv("https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/2nd_challenge/raw_datasets/prediction_data/2022BD_subject.tsv", sep='\t')
#new = pd.merge(df_2022_specimen,df_2022_subject, on= 'subject_id')
new = pd.concat([pd.merge(df_2020_specimen,df_2020_subject, on= 'subject_id'), 
                 pd.merge(df_2021_specimen,df_2021_subject, on= 'subject_id')])


titer_df = new.merge(titer_df, on='specimen_id')
titer_df_filtered = titer_df[(titer_df['isotype'] == 'IgG') & (titer_df['antigen'] == 'PT') & (titer_df['is_antigen_specific'] == True)]

cell_freq = cell_freq[(cell_freq['cell_type_name'] == 'Monocytes')]
cell_df = new.merge(cell_freq, on='specimen_id')

gene = gene[gene['versioned_ensembl_gene_id'] == 'ENSG00000277632.1']
gene_df = new.merge(gene, on='specimen_id')

df_model_evaluation = pd.read_csv('model_evaluation_task11.csv')
df_r2_scores = pd.read_csv('model_r2_scores.csv')
```

```python


def generate_age_distribution(df):
    fig = px.histogram(df, x='Age', nbins=20, title="Age Distribution")
    return fig

def generate_sex_distribution(df):
    sex_distribution = df['biological_sex_Male'].map({1.0: 'Male', 0.0: 'Female'})
    fig = px.pie(names=sex_distribution, title="Sex Distribution")
    return fig


app = dash.Dash(__name__)
app = dash.Dash(external_stylesheets=[dbc.themes.MINTY])

lasso = dbc.Card(dcc.Graph(id='fig', style={'padding': 10, 'flex': 1, "height": "50vh"}))
rf = dbc.Card(dcc.Graph(id='rf', style={'padding': 10, 'flex': 1, "height": "50vh"}))
r2 = dbc.Card(dcc.Graph(id='r2', style={'padding': 10, 'flex': 1, "height": "50vh"}))
imm = dbc.Card(dcc.Graph(id='imm', style={'padding': 10, 'flex': 1, "height": "50vh"}))
eva = dbc.Card(dcc.Graph(id='eval', style={'padding': 10, 'flex': 1, "height": "50vh"}))
res = dbc.Card(dcc.Graph(id='results', style={'padding': 10, 'flex': 1, "height": "50vh"}))

app.layout = html.Div([
    html.H1("Immune Responses to Vaccination Dashboard"),
    html.Div([drops]
            ),
    html.Div([
        html.Div([
            dcc.Graph(
                figure=generate_age_distribution(tpf.df_train).update_layout(template='xgridoff')
            )
        ], style={'display': 'inline-block', 'width': '50%'}),
        
        html.Div([
            dcc.Graph(
                figure=generate_sex_distribution(tpf.df_train).update_layout(template='ggplot2')
            )
        ], style={'display': 'inline-block', 'width': '50%'})
    ], style={'width': '100%', 'display': 'flex'}),
    
    html.Div([
        html.Div([
            lasso
        ], style={'display': 'inline-block', 'width': '50%'}),
        
        html.Div([
            rf
        ], style={'display': 'inline-block', 'width': '50%'})
    ], style={'width': '100%', 'display': 'flex'}),
    
    html.Div([
        html.Div([
            r2
        ], style={'display': 'inline-block', 'width': '50%'}),
        
        html.Div([
            imm
        ], style={'display': 'inline-block', 'width': '50%'})
    ], style={'width': '100%', 'display': 'flex'}),
    
    html.Div([
        html.Div([
            eva
        ], style={'display': 'inline-block', 'width': '100%'}),
    ], style={'width': '100%', 'display': 'flex'}),
    
    html.Div([
        html.Div([
            dash_table.DataTable(
        id='table',
        columns=[{"name": i, "id": i} for i in tpf.df_task1.columns],
        data=tpf.df_task1.to_dict('records'),
        style_table={'height': '300px', 'overflowY': 'auto'},
        page_size=10, 
    ),
        ], style={'display': 'inline-block', 'width': '100%'}),
    ], style={'width': '100%', 'display': 'flex'}),
    
])

@app.callback(
    dash.dependencies.Output('fig', 'figure'),
    [dash.dependencies.Input('demo-dropdown', 'value')])
def generate_lasso_coefficients_plot(value):
    if value == 'Antibody Titers':
        X_columns = tpf.X_train_scaled.columns
        coefficients = tpf.lasso_coefficients
        df = pd.DataFrame({
            'Feature': X_columns,
            'Absolute Coefficient': abs(coefficients)
        }).sort_values(by='Absolute Coefficient')

        fig = px.bar(df, y='Feature', x='Absolute Coefficient', orientation='h',
                     title='Lasso Coefficients (Absolute Values)')
    elif value == 'Monocytes':
        X_columns = tp2.X_train_scaled.columns
        coefficients = tp2.lasso_coefficients
        df = pd.DataFrame({
            'Feature': X_columns,
            'Absolute Coefficient': abs(coefficients)
        }).sort_values(by='Absolute Coefficient')

        fig = px.bar(df, y='Feature', x='Absolute Coefficient', orientation='h',
                     title='Lasso Coefficients (Absolute Values)')
    elif value == 'CCL3':
        X_columns = tp3.X_train_scaled.columns
        coefficients = tp3.lasso_coefficients
        df = pd.DataFrame({
            'Feature': X_columns,
            'Absolute Coefficient': abs(coefficients)
        }).sort_values(by='Absolute Coefficient')

        fig = px.bar(df, y='Feature', x='Absolute Coefficient', orientation='h',
                     title='Lasso Coefficients (Absolute Values)')
    return fig

@app.callback(
    dash.dependencies.Output('rf', 'figure'),
    [dash.dependencies.Input('demo-dropdown', 'value')])
def generate_rf_feature_importances_plot(value):
    if value == 'Antibody Titers':
        X_columns = tpf.X_train_scaled.columns
        importances = tpf.feature_importances
        df = pd.DataFrame({
            'Feature': X_columns,
            'Feature Importance': importances
        }).sort_values(by='Feature Importance')

        fig = px.bar(df, y='Feature', x='Feature Importance', orientation='h',
                     title='Random Forest Feature Importances')
    elif value == 'Monocytes':
        X_columns = tp2.X_train_scaled.columns
        importances = tp2.feature_importances
        df = pd.DataFrame({
            'Feature': X_columns,
            'Feature Importance': importances
        }).sort_values(by='Feature Importance')

        fig = px.bar(df, y='Feature', x='Feature Importance', orientation='h',
                     title='Random Forest Feature Importances')
    elif value == 'CCL3':
        X_columns = tp3.X_train_scaled.columns
        importances = tp3.feature_importances
        df = pd.DataFrame({
            'Feature': X_columns,
            'Feature Importance': importances
        }).sort_values(by='Feature Importance')

        fig = px.bar(df, y='Feature', x='Feature Importance', orientation='h',
                     title='Random Forest Feature Importances')
        
    return fig

@app.callback(
    dash.dependencies.Output('r2', 'figure'),
    [dash.dependencies.Input('demo-dropdown', 'value')])
def generate_model_evaluation_chart(value):
    if value == 'Antibody Titers':
        r2_dict = tpf.r2_dict
        models = list(r2_dict.keys())
        scores = list(r2_dict.values())
        fig = px.bar(x=models, y=scores, labels={'x': 'Model', 'y': 'R-squared'}, title="Model Evaluation R-squared").update_layout(template='xgridoff')
    if value == 'Monocytes':
        r2_dict = tp2.r2_dict
        models = list(r2_dict.keys())
        scores = list(r2_dict.values())
        fig = px.bar(x=models, y=scores, labels={'x': 'Model', 'y': 'R-squared'}, title="Model Evaluation R-squared").update_layout(template='xgridoff')
    if value == 'CCL3':
        r2_dict = tp3.r2_dict
        models = list(r2_dict.keys())
        scores = list(r2_dict.values())
        fig = px.bar(x=models, y=scores, labels={'x': 'Model', 'y': 'R-squared'}, title="Model Evaluation R-squared").update_layout(template='xgridoff')
    return fig

@app.callback(
    dash.dependencies.Output('imm', 'figure'),
    [dash.dependencies.Input('demo-dropdown', 'value')])
def generate_immunogenicity_plot(value):
    if value == 'Antibody Titers':
        df = titer_df_filtered
        y_axis = 'MFI_normalised'
        title = 'MFI Normalised'
    elif value == 'Monocytes':
        df = cell_df
        y_axis = 'percent_live_cell'
        title = 'Percent Live Cell'
    elif value == 'CCL3':
        df = gene_df
        y_axis = 'tpm'
        title = 'TPM'
    df_filtered = df[(df['actual_day_relative_to_boost'] >= df['actual_day_relative_to_boost'].min()) & 
                     (df['actual_day_relative_to_boost'] <= 150)]
    
    fig = px.line(
        df_filtered,
        x='actual_day_relative_to_boost',
        y=y_axis,
        color='infancy_vac',
        line_group='subject_id',
        markers=True,
        title='Immunogenicity Over Time'
    )
    
    fig.update_xaxes(range=[df_filtered['actual_day_relative_to_boost'].min(), 150])

    fig.add_vline(x=0, line=dict(color="black", dash="dash"))
    fig.add_vline(x=14, line=dict(color="black", dash="dash"))

    fig.update_layout(
        xaxis_title="Actual Day Relative to Boost",
        yaxis_title= title,
        template="simple_white",
        legend_title_text='Infancy Vaccine'
    )
    fig.update_layout(template='presentation')
    return fig


@app.callback(
    dash.dependencies.Output('eval', 'figure'),
    [dash.dependencies.Input('demo-dropdown', 'value')])
def generate_bar_graph(value):
    df = df_r2_scores
    if value == 'Antibody Titers':
        cols = ["1.1", "1.2"]
        title = "Comparison of 1.1 and 1.2 Metrics"
    elif value == 'Monocytes':
        cols = ["2.1", "2.2"]
        title = "Comparison of 2.1 and 2.2 Metrics"
    elif value == 'CCL3':
        cols = ["3.1", "3.2"]
        title = "Comparison of 3.1 and 3.2 Metrics"
    melted_df = df.melt(id_vars=["Task"], value_vars=cols, var_name="Metric", value_name="Score")
    fig = px.bar(
        melted_df,
        x='Task',
        y='Score',
        color='Metric',
        barmode='group',
        title=title
    )
    return fig

@app.callback(
    dash.dependencies.Output('results', 'figure'),
    [dash.dependencies.Input('demo-dropdown', 'value')])
def generate_results(value):
    df = tpf.df_task1
    fin = dash_table.DataTable(
        id='table',
        columns=[{"name": i, "id": i} for i in tpf.df_task1.columns],
        data=tpf.df_task1.to_dict('records'),
        style_table={'height': '300px', 'overflowY': 'auto'},
        page_size=10, 
    )
    return fin

if __name__ == '__main__':
    app.server.run(port=8000, host='127.0.0.1')

```
